# Sales Price

A local department store needs to develop a program that will computer the total price (including goods and services tax) of the item on sale when given an item's original price and discount percentage.

## Tasks
* Save to Github
* Design Algorithm
* Write and test the program in Python

```python
def calculate_total_price(original_price,discount_percentage,tax_percentage):
  discount=original_price * discount_percentage/100
  selling_price=original_price  -discount
  GST=selling_price * tax_percentage/100
  total_price=selling_price + GST

  return total_price

original_price=input("Enter your price: ")
discount_percentage=input("Enter your percentage: ")
tax_percentage=input("Enter your percentage: ")

x=int(original_price)
y=int(discount_percentage)
z=int(tax_percentage)

total_price=calculate_total_price(x,y,z)
print(total_price)

```
